```python
import pandas as pd
```

```python
census_df = pd.read_csv('Raw Data/raw_data.csv')
census_df.head()
```

```python
# Split the name column to break up the State and County name
name_split = census_df['NAME'].str.split(',', expand=True)

# Rename the columns and reorder so state goes before county
name_split.rename(columns={0:'County',1:'State'},inplace=True)
reordered_columns = name_split[['State','County']]

# Add the new columns to the census_df and drop the old columns
census_df = pd.concat([reordered_columns, census_df], axis=1)
census_df = census_df.drop(columns=['Unnamed: 0','NAME'])
census_df.head()

# Move Geo_ID to the front
column_to_move = 'GEO_ID'
moved_column = census_df.pop(column_to_move)
census_df.insert(0,column_to_move,moved_column)
census_df.head()
```

```python
# We want to remove the margin of error columns and just focus on the census estimate amounts
# Create a list of columns that end with 'M'
margin_columns = [col for col in census_df.columns if col.endswith('M')]

# Drop the margin columns from census_df
census_df = census_df.drop(margin_columns, axis=1)
census_df.head()
```
